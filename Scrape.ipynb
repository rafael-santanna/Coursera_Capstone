```python
import pandas as pd
import numpy as np
```

```python
data_csv = {'Postal Code': [], 'Borough': [], 'Neighbourhood': []}
keys = ['Postal Code', 'Borough', 'Neighbourhood']
test = 's'
with open('data.txt', 'r') as data:
    while True:
        if test == '': break
        count = 0
        while count < 3:
            line = data.readline()
            data_csv[keys[count]].append(line[2:].strip())
            count += 1
        test = data.readline()
```

```python
df = pd.DataFrame(data_csv)
df
```

```python
df.drop(df[df['Borough']=='Not assigned'].index, inplace=True)
df.reset_index(inplace=True)
df.head(12)
```
